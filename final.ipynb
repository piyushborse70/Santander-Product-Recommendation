```python
import json

def Read_Probabilities():
  '''
  This Function Returns the Dictionary of Probabilities for each id
  '''
  with open('/content/drive/MyDrive/Colab Notebooks/Applied Ai/PROJECT/Predictions.txt') as json_file:
    ProbDict = json.load(json_file)

  ProbDict = {int(k):v for k,v in ProbDict.items()}
  return ProbDict

def recommend_Product(id):
  '''
  This Function Returns Recommendation of Products for Given ID
  '''
  Prob = Read_Probabilities()
  return Prob[id]
```
